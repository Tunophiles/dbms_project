```python
import os
import csv

# Step 1: List the songs in a directory
directory = "C:/Users/pavan/Downloads/[SPOTIFY-DOWNLOADER.COM] Hot Hits Hindi"
songs = os.listdir(directory)

# Step 2: Create and write the songs to a CSV file
csv_file = "songs_mf.csv"  # Name of the CSV file

with open(csv_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    
    # Write a header row
    writer.writerow(["Song Name"])
    
    # Write the list of songs to the CSV file
    for song in songs:
        writer.writerow([song])

print(f"Songs list has been written to {csv_file}")
```

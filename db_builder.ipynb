```python
song_df = pd.read_csv('final_data.csv')
```

```python
song_df = song_df[['id','name',"artists",'release_year','duration_ms']]
                  
                  
song_df.columns[:30]
```

```python
# song_df.head()
song_df = song_df[~song_df.isnull()['name']]
```

```python
artists = song_df['artists'].to_list()
```

```python
art = [i[1:-1].split(', ') for i in artists]
# art = [i[1:-1] for i in art]
artists_list = []
for i in art:
    x = []
    for j in i:
        x.append(j[1:-1])
    artists_list.append(x)
artists_list[:5]
```

```python
song_df.drop('artists', axis=1, inplace=True)
```
